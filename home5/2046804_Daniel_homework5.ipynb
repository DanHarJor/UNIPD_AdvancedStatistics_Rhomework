## uniform prior

```r
data <- c(4,1,3,1,3)
u <- seq(0,7,0.1)
# uniform prior from 0 to 10 assumed for u 
likelihood <- function(u){
    poisson <- vector()
    for(data_i in data){
        poisson <- c(poisson,dpois(data_i,u))
    }
    return(prod(poisson))
}

normC <- integrate(Vectorize(likelihood), lower = 0, upper = max(u))$value

posterior <- function(u){
    return(likelihood(u)/normC)
}



cat('The area under the posterior should be 1, it is: ', integrate(Vectorize(posterior),lower = 0, upper = max(u))$value)

posterior_v <- vector()
for(u_i in u){
    posterior_v = c(posterior_v,posterior(u_i))
}

alpha <- sum(data)+1
beta <- length(data)

gamma <- dgamma(u, alpha, beta)

#analytical
mean_a <- alpha/beta
var_a <- alpha/(beta^2)

cat("\nThe analytical mean and variance is: ","mean: ",mean_a," variance: ",var_a)

uP <- function(u){
    u*posterior(u)
}

u2P <-function(u){
    u^2*posterior(u)
}

mean_n <- integrate(Vectorize(uP),lower = 0, upper = max(u))$value 
var_n <- abs(mean_n^2 - integrate(Vectorize(u2P),lower = 0, upper = max(u))$value)

cat("\nThe numerical mean and variance is: ","mean: ",mean_n," variance: ",var_n)

median <- qgamma(0.5,alpha,beta)
cat("\nThe median can only be calculated numerically, ",median)

plot(u,gamma,lwd = 3, type = 'l',col = 'blue',main = 'Poisson Posterior with uniform Prior',
    xlab = 'poisson frequency, u',ylab = 'probability of u given the data')
abline(v = median,col = 'blue',lwd = 3)
abline(v = qgamma(0.025,alpha,beta),col = 'blue')
abline(v = qgamma(0.975,alpha,beta),col = 'blue')

lines(u,dnorm(u,mean_a,sqrt(var_a)), col = 'red',lwd = 3)
abline(v = mean_a, col = 'red',lwd = 3)
abline(v = qnorm(0.025,mean_a,sqrt(var_a)),col = 'red')
abline(v = qnorm(0.975,mean_a,sqrt(var_a)),col = 'red')

legend("topright", 
       inset=0.1, 
       legend=c("Posterior","Normal"),
       col=c("Blue", "red"), 
       lty=1:1, 
       cex=1,
       box.lty=1,
       lwd = 4)

cat('\n\nIn the graph the posterior is shown with the equilivant normal distribution of same mean and variance.')
cat('\nThe central vertical bars are blue for posterior median and red for shared mean.')
cat('\nThe external vertical bars show the 95% confidance intervals')
```

## jeffrey's prior

```r
jeff <- function(u){
    return(1/(u^2))
}

likelihood_j <- function(u){
        return(jeff(u)*likelihood(u))
}

normC <- integrate(Vectorize(likelihood_j), lower = 0, upper = max(u))$value

posterior_j <- function(u){
    return(likelihood_j(u)/normC)
}

# plus 0.5 rather than one since its jeffry's prior
alpha <- sum(data)+0.5
beta <- length(data)

gamma <- dgamma(u, alpha, beta)

#analytical
mean_a <- alpha/beta
var_a <- alpha/(beta^2)

cat("\nThe analytical mean and variance is: ","mean: ",mean_a," variance: ",var_a)

uP <- function(u){
    u*posterior_j(u)
}

u2P <-function(u){
    u^2*posterior_j(u)
}

mean_n <- integrate(Vectorize(uP),lower = 0, upper = max(u))$value 
var_n <- abs(mean_n^2 - integrate(Vectorize(u2P),lower = 0, upper = max(u))$value)

cat("\nThe numerical mean and variance is: ","mean: ",mean_n," variance: ",var_n)

median <- qgamma(0.5,alpha,beta)
cat("\nThe median can only be calculated numerically, ",median)

plot(u,gamma,lwd = 3, type = 'l',col = 'blue',main = "Poisson Posterior with Jeffry's Prior",
    xlab = 'poisson frequency, u',ylab = 'probability of u given the data')
abline(v = median,col = 'blue',lwd = 3)
abline(v = qgamma(0.025,alpha,beta),col = 'blue')
abline(v = qgamma(0.975,alpha,beta),col = 'blue')

lines(u,dnorm(u,mean_a,sqrt(var_a)), col = 'red',lwd = 3)
abline(v = mean_a, col = 'red',lwd = 3)
abline(v = qnorm(0.025,mean_a,sqrt(var_a)),col = 'red')
abline(v = qnorm(0.975,mean_a,sqrt(var_a)),col = 'red')

legend("topright", 
       inset=0.1, 
       legend=c("Posterior","Normal"),
       col=c("Blue", "red"), 
       lty=1:1, 
       cex=1,
       box.lty=1,
       lwd = 4)

cat('\n\nIn the graph the posterior is shown with the equilivant normal distribution of same mean and variance.')
cat('\nThe central vertical bars are blue for posterior median and red for shared mean.')
cat('\nThe external vertical bars show the 95% confidance intervals')
```

# Exercise 2

light house problem

As taken from the lecture slides the likelihood is given by

$$P(x \mid \alpha, \beta)=\frac{1}{\pi} \frac{\beta}{\beta^{2}+(x-\alpha)^{2}}$$

Assuming we know nothing of the situation a uniform prior is selected for both alpha and beta.

This means the Posterior is a normalised version of the likelihood.


```r
#this 3d plotting code was borrowed from https://stackoverflow.com/questions/30960620/how-to-plot-a-plane-from-an-equation-in-r
x <- y <- seq(-10, 10, length= 30)
f <- function(x,y){ z <- x*2 + y -3 }
z <- outer(x,y,f)
persp(x, y, z, theta = 30, phi = 30, expand = 0.5, col = "lightblue")
```

```r
#x = 3
likelihood <- function(alpha, beta){
    return((1/pi)  *  (beta / (beta^2 + (3-alpha)^2)))
}

#trial some range of alpha and beta
x <- seq(-6,6,3)
alpha <- seq(-20,20,1)
beta <- seq(0,20,1)

likelihood_plane <- outer(alpha,beta,likelihood)
persp(alpha,beta,likelihood_plane, theta = 40, phi = 30, main = 'Likelihood', col = 'lightblue',axes=T, ticktype="detailed")


likelihoods <- vector()
for(alpha_i in alpha){
    for(beta_i in beta){
        likelihoods <- c(likelihoods, likelihood(alpha_i, beta_i))
        #print(likelihood(alpha_i, beta_i))
    }
}

normC <- sum(likelihoods,na.rm=T)
cat('for x = 3 the sum of the likelihoods is', normC)
print('this can be used as the normalisation coefficiant for the posterior')

posterior <- function(alpha,beta){likelihood(alpha,beta)/normC}
posterior_plane <- outer(alpha,beta,posterior)
persp(alpha,beta,posterior_plane, theta = 40, phi = 30, main = 'Posterior', col = 'lightblue',axes=T, ticktype="detailed")

```

# Generate the Signal and Calculate the Posterior
## altering sampling resolution w

```r
A.true <- 2 # Signal amplitude
B.true <- 1 # Background amplitude
Delta.t <- 5 # Exposure time

signal_posterior <- function(w,A.true,B.true){
    # - Generative model
    signal <- function(x, a, b, x0, w, t) {
    t * (a*exp(-(x-x0)^2/(2*w^2)) + b)
    }
    # Define model parameters
    x0 <- 0 # Signal peak

    
    # - Generate the observed data
    set.seed(205)
    xdat <- seq(from=-7*w, to=7*w, by=0.5*w)
    s.true <- signal(xdat , A.true , B.true , x0, w, Delta.t)
    ddat <- rpois(length(s.true), s.true)
    xplot <- seq(from=min(xdat), to=max(xdat), by=0.05*w)
    splot <- signal(xplot , A.true , B.true , x0, w, Delta.t)
    plot(xplot , splot ,
    xlab="x", ylab="Signal+Background counts")
    par(new=TRUE)
    xdat.off <- xdat -0.25
    sigtitle = paste('Signal, Sampling Resolution: ',w,"Signal to Noise Ratio: ", A.true/B.true)
    plot(main = sigtitle,xdat.off, ddat , type='s',col='firebrick3',
    lwd=2,xlim=range(xplot), ylim=range(c(splot , ddat )))
    
    
    alim <- c(0.0, 4.0)
    blim <- c(0.5, 1.5)
    Nsamp <- 100
    uniGrid <- seq(from=1/(2*Nsamp),
    to=1-1/(2*Nsamp), by=1/Nsamp)
    delta_a <- diff(alim )/ Nsamp
    delta_b <- diff(blim )/ Nsamp
    a <- alim[1] + diff(alim )* uniGrid
    b <- blim[1] + diff(blim )* uniGrid

    # Log posterior
    log.post <- function(d, x, a, b, x0, w, t) {
    if(a<0 || b <0) {return(-Inf )} # the effect of the prior
    sum(dpois(d, lambda=signal(x, a, b, x0, w, t), log=TRUE))
    }
    # Compute log unnormalized posterior , z = ln Pˆ*(a,b|D), on a regular grid
    z <- matrix(data=NA , nrow=length(a), ncol=length(b))
    for(j in 1:length(a)) {
    for(k in 1:length(b)) {
    z[j,k] <- log.post(ddat , xdat , a[j], b[k], x0, w, Delta.t)
    }
    }
    z <- z - max(z) # set maximum to zero
    # Plot unnormalized 2D posterior as contours.
    potitle = paste('Posterior, Sampling Resolution: ',w,"Signal to Noise Ratio: ", A.true/B.true)
    contour(a, b, exp(z), main = potitle,
    nlevels = 5,
    #labcex = 0.5,
    #lwd = 2,
    xlab="amplitude , A",
    ylab="background , B")
    abline(v=2,h=1,col="grey")
}

w <- c(0.1, 0.25, 1, 2, 3)

for(w_i in w){
    signal_posterior(w_i,A.true,B.true)
}
```

# Altering the A/B ratio

```r

A.true <- c(0.5,1,2,3,4)
w <- 1
for(A.true_i in A.true){
    signal_posterior(w, A.true_i, B.true)
}
```
